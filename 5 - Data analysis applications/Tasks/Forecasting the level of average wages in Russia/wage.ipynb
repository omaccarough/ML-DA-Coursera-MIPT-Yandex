# Прогнозирование среднего уровня заработной платы в России.

Известны данные о заработной плате за каждый месяц с января 1993 по август 2016.  
Необходимо проанализировать данные, подобрать для них оптимальную прогнозирующую модель в классе ARIMA и построить прогноз на каждый месяц на два года вперёд от конца данных.

```python
%matplotlib inline
import warnings
import datetime
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from scipy import stats
from itertools import product
from dateutil.relativedelta import relativedelta

def invboxcox(y,lmbda):  # Обратное преобразование Бокса-Кокса.
   if lmbda == 0:
      return(np.exp(y))
   else:
      return(np.exp(np.log(lmbda*y+1)/lmbda))
```

## Визуальный анализ ряда.

```python
wage = pd.read_csv('WAG_C_M.csv', sep=';', index_col=['month'],
                   parse_dates=['month'], dayfirst=True)
wage.rename(columns={'WAG_C_M': 'wage'}, inplace=True)

plt.figure(figsize=(15,8))
plt.plot(wage.wage, color='b')
plt.title('Средняя заработная плата с января 1993 по август 2016',
         fontsize=15)
plt.xlabel('Month', fontsize=13)
plt.ylabel('Wage', fontsize=13)
```

При первичном визуальном анализе ряда можно сделать некоторые выводы:  
1. Ряд имеет ярковыраженный повышающийся тренд.
2. В ряде достаточно сильно выражена годовая сезонность.
3. В ряду со временем изменяется дисперсия: размах сезонных колебаний в начале сильно отличается от сезонных колебаний, находящихся в конце ряда.

### Критерий Дики-Фулера.

```python
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(wage.wage)[1])
```

Достигаемый уровень значимости критерия Дики-Фулера намного больше 0.05, следовательно гипотеза о нестационарности ряда не отвергается.

###  STL-декомпозиция.

```python
res = sm.tsa.seasonal_decompose(wage.wage)

fig, (ax0, ax1,ax2,ax3) = plt.subplots(4,1, figsize=(15,10))
ax0.plot(wage.wage, color='b')
ax0.set_ylabel('Observed')
res.trend.plot(ax=ax1, color='b')
ax1.set_ylabel('Trend', fontsize=13)
res.seasonal.plot(ax=ax2, color='b')
ax2.set_ylabel('Seasonal', fontsize=13)
res.resid.plot(ax=ax3, color='b')
ax3.set_ylabel('Residual', fontsize=13)
```

После проведения STL-декомпозиции ряда можно сделать следующие выводы:  
1. Достаточно хорошо выраженный повышающийся тренд.  
2. Также хорошо выражена сезонность этого ряда (период составляет 12 месяцев).  
3. Дисперсия значений ряда в начале значительно меньше значений ряда в конце.
4. В остатках прослеживается структура.

Так как в ряду изменяется дисперсия и наблюдается тренд с сезонностью, нужно провести стабилизацию дисперсии и сделать дифференцирование.

## Стабилизация дисперсии.

Сделаем преобразование Бокса-Кокса для стабилизации дисперсии.

```python
wage['wage_box'], lmbda = stats.boxcox(wage.wage)
plt.figure(figsize=(15,8))
plt.plot(wage.wage_box, color='b')
plt.xlabel('Month', fontsize=13)
plt.ylabel('Transformed wage', fontsize=13)

print("Оптимальный параметр преобразования Бокса-Кокса: %f" % lmbda)
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(wage.wage_box)[1])
```

По графику видно, что после применения преобразования Бокса-Кокса дисперсия уменьшилась. Но по критерию Дики-Фулера нулевая гипотеза о нестационарности ряда не отвергается (0.697 > 0.05). Также до сих пор наблюдаются тренд и сезонность. Следовательно, стационарность не была достигнута.

## Выбор порядка дифференцирования.

### Сезонное дифференцирование.

Так как у ряда присутствует ярко выраженный сезонный профиль, то сначала будет применено сезонное дифференцирование. Далее выполнены STL-декомпозиция ряда и проверка стационарности.

```python
wage['wage_box_diff'] = wage.wage_box - wage.wage_box.shift(12)
res2 = sm.tsa.seasonal_decompose(wage.wage_box_diff[12:])

fig, (ax0, ax1,ax2,ax3) = plt.subplots(4,1, figsize=(15,10))
ax0.plot(wage.wage_box_diff, color='b')
ax0.set_ylabel('Observed')
res2.trend.plot(ax=ax1, color='b')
ax1.set_ylabel('Trend', fontsize=13)
res2.seasonal.plot(ax=ax2, color='b')
ax2.set_ylabel('Seasonal', fontsize=13)
res2.resid.plot(ax=ax3, color='b')
ax3.set_ylabel('Residual', fontsize=13)

print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(wage.wage_box_diff[12:])[1])
```

Достигаемый уровень значимости критерия Дики-Фулера позволяет нам отвергнуть нулевую гипотезу о нестационарности. Но глядя на графики временного ряда и тренда можно заметить, что полностью избавиться от тренда не удалось. Следовательно, стационарным этот ряд считать нельзя. Поэтому будет проведено обычное дифференцирование.

### Обычное дифференцирование.

```python
wage['wage_box_diff2'] = wage.wage_box_diff - wage.wage_box_diff.shift(1)
res3 = sm.tsa.seasonal_decompose(wage.wage_box_diff2[13:])

fig, (ax0, ax1,ax2,ax3) = plt.subplots(4,1, figsize=(15,10))
ax0.plot(wage.wage_box_diff2, color='b')
ax0.set_ylabel('Observed')
res3.trend.plot(ax=ax1, color='b')
ax1.set_ylabel('Trend', fontsize=13)
res3.seasonal.plot(ax=ax2, color='b')
ax2.set_ylabel('Seasonal', fontsize=13)
res3.resid.plot(ax=ax3, color='b')
ax3.set_ylabel('Residual', fontsize=13)

print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(wage.wage_box_diff2[13:])[1])
```

Достигаемый уровень значимости критерия Дики-Фулера на продифференцированном ряду позволяет отвергнуть нулевую гипотезу о нестационарности. Также, больше не наблюдается никакого систематического поведения в трендовой компоненте - она колеблется примерно вокруг константы.

## Выбор начальных приближений для p, q, P, Q.

Посмотрим на ACF и PACF полученного ряда:

```python
plt.figure(figsize=(15,8))
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(wage.wage_box_diff2[13:].values.squeeze(), lags=48, ax=ax)
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(wage.wage_box_diff2[13:].values.squeeze(), lags=48, ax=ax)
plt.show()
```

### Начальные приближения:
Q*S - номер последнего сезонного лага при котором автокорреляция значима. По графику видно, что значимых корреляций с номером сезонного лага кратному 12 (S=12) не наблюдается. Следовательно, Q=0.   

q - номер последнего несезонного лага, при котором автокорреляция значима. q=1.

P*S -  номер последнего сезонного лага при котором частичная автокорреляция значима. По графику видно, что это лаг под номером 48. Следовательно, P=4.

p - номер последнего несезонного лага при котором частичная автокорреляция значима. p=10.

Q=0, q=1, P=4, p=10.

## Обучение и сравнение моделей.

```python
D=1
d=1
Qs = range(0, 1)
qs = range(0, 2)
Ps = range(0, 5)
ps = range(0, 11)
```

```python
parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
len(parameters_list)
```

```python
%%time
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')

for param in parameters_list:
    try:
        model=sm.tsa.statespace.SARIMAX(wage.wage_box, order=(param[0], d, param[1]), 
                                        seasonal_order=(param[2], D, param[3], 12)).fit(disp=-1)
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
    
warnings.filterwarnings('default')
```

```python
# Пять лучших моделей.

result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
```

```python
# Лучшая модель.

print(best_model.summary())
```

##  Анализ остатков построенной модели.

```python
plt.figure(figsize=(15,8))
plt.subplot(211)
best_model.resid[13:].plot(color='b')
plt.ylabel('Residuals')

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(best_model.resid[13:].values.squeeze(), lags=48, ax=ax)

print("Критерий Стьюдента: p=%f" % stats.ttest_1samp(best_model.resid[13:], 0)[1])
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(best_model.resid[13:])[1])
```

При визуальном анализе остатков можно отметить, что в них не наблюдаются тренд и сезонность. На коррелограмме отстутствует структура.


Проверка свойств остатков с помощью формальных критериев:  

Критерий Стьюдента: p-value=0.293 > 0.05 - остатки несмещены.  
Критерий Дики-Фуллера: p-value < 0.05 - остатки стационарны.  
Критерий Льюнга-Бокса: p-value=0.37 > 0.05 - остатки неавтокоррелированы.

### Приближение данных моделью:

```python
wage['model'] = invboxcox(best_model.fittedvalues, lmbda)
plt.figure(figsize=(15,7))
wage.wage.plot(color='b')
wage.model[13:].plot(color='r')
plt.ylabel('Wage', fontsize=13)
```

При визуальном анализе можно отметить, что модель и данные очень похожи друг на друга.

## Прогнозирование с помощью финальной модели.

```python
wage2 = wage[['wage']]
date_list = [datetime.datetime.strptime("2016-08-01", "%Y-%m-%d") + relativedelta(months=x) for x in range(0,36)]
future = pd.DataFrame(index=date_list, columns= wage2.columns)
wage2 = pd.concat([wage2, future])
wage2['forecast'] = invboxcox(best_model.predict(start=283, end=320), lmbda)

plt.figure(figsize=(15,7))
wage2.wage.plot(color='b', label='Фактические значения')
wage2.forecast.plot(color='r', label='Предсказанные значения')
plt.ylabel('Wage', fontsize=13)
plt.title('Прогноз средней заработной платы с августа 2016 по август 2019', fontsize=13)
plt.legend()
```

Прогноз выглядит достаточно адекватным. В нём присутствуют как возрастающий тренд, так и сезонные колебания.
